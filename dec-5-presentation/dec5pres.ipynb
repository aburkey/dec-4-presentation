```python
import matplotlib
import matplotlib.pyplot as plt
```

Adding text to plotting  Matplotlib can be done through a variety of commands. You can change the height of your graph with the fig.subplots_adjust command, and changing the top value to your desired scale. Use the ax.axis([]) command. 

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=.85)
ax.axis([0, 10, 0, 10])
```

You can add a chart title with the fig.suptitle command and customize the font size and weight, and axis titles and labels with ax.set_ command. 

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=.85)
ax.axis([0, 10, 0, 10])

fig.suptitle('chart title', fontsize=14, fontweight='bold')
ax.set_title('axes title')
ax.set_xlabel('x axis label', fontsize=12)
ax.set_ylabel('y axis label', fontsize=12)
```

There are lots of fun additions you can make to your chart or graph using some simple commands from matplotlib. You can add a color fill to a piece of boxed text, and you can even change the font style. Designate values to 'alpha' for the opacity of the colorfill and 'pad' for the amount of space around teh text to be enclosed in the box. You can add other text lines inside your chart or graph in specific places using the coordinates of the place you would like inside the ax.text() command. If there is not a specific coordinate you would like you text in, you can simply write the position within the chart where you would like the text. Finally, you can create and annotate a point with text or symbols like arrows. Notice specific syntax and indentation needed for each command. the dict() function can determine the color of the symbol along with the space between the point and the symbol. 

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=0.85)
ax.axis([0, 10, 0, 10])

fig.suptitle('chart title', fontsize=14, fontweight='bold')
ax.set_title('axes title')
ax.set_xlabel('x axis label', fontsize=12)
ax.set_ylabel('y axis label', fontsize=12)

ax.text(3, 8, 'boxed text', style='italic',
        bbox={'facecolor': 'pink', 'alpha': .9, 'pad': 12})

ax.text(2, 6, r'an equation: $E=mc^2$', fontsize=15)

ax.text(3, 2, 'text inside graph')

ax.text(0.95, 0.01, 'colored text inside graph',
        verticalalignment='bottom', horizontalalignment='right',
        transform=ax.transAxes,
        color='purple', fontsize=15)

ax.plot([2], [1], 'o')
ax.annotate('annotate a point', xy=(2, 1), xytext=(3, 4),
            arrowprops=dict(facecolor='blue', shrink=0.06))
plt.show()
```

The positions of the x and y axes can be shifted, too,  using the labelpad function. In this example, we shifted the y axis label further left from the graph, and the x axis further down from the graph. If we want to shift the label to the right or left, we need to input a specific point. 

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=0.85)
ax.axis([0, 10, 0, 10])

fig.suptitle('chart title', fontsize=14, fontweight='bold')
ax.set_title('axes title')
ax.set_xlabel('x axis label', fontsize=15)
ax.set_ylabel('y axis label', fontsize=15)


ax.set_xlabel('x axis label', labelpad=20, position=(0., 1e6), horizontalalignment='left')
ax.set_ylabel('y axis label]', labelpad=20)
plt.show()
```

You can also change the font with matplotlib!

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=0.85)
ax.axis([0, 10, 0, 10])

fig.suptitle('chart title', fontsize=14, fontweight='bold')
ax.set_title('axes title')
ax.set_xlabel('x axis label', fontsize=14)
ax.set_ylabel('y axis label', fontsize=14)

ax.set_xlabel('x axis label', labelpad=20, position=(0., 1e6), horizontalalignment='left')
ax.set_ylabel('y axis label]', labelpad=20)

from matplotlib.font_manager import FontProperties
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')
font.set_style('italic')


plt.show()
```

We can change the position (vertically and horizontally) of chart titles, also, using the loc argument. 

```python
fig = plt.figure()
ax = fig.add_subplot()
fig.subplots_adjust(top=0.85)
ax.axis([0, 10, 0, 10])

fig.suptitle('chart title', fontsize=14, fontweight='bold')
ax.set_title('axes title')
ax.set_xlabel('x axis label', fontsize=14)
ax.set_ylabel('y axis label', fontsize=14)


loc = ['center', 'left', 'right']
ax.set_title('axes title '+loc, loc=loc)
plt.show()

```

Finally, we will cover how to input changes to your tick marks on a chart or graph. We can change the value of the x-axis tick marks from the default setting using the set_ticks argument

We can also use list comprehensions to change the tick labels. 

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
ticks = np.arange(0., 8.1, 2.)

tickla = [f'{tick:1.2f}' for tick in ticks]
axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_ticklabels(tickla)
axs[1].set_xlim(axs[0].get_xlim())
plt.show()
```

Instead of making a list of all of the ticklabels, though, we could use strings to dictate the ticks along with the amount of decimal points wanted. 

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
ticks = np.arange(0., 8.1, 2.)
axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_major_formatter('{x:1.1f}')
axs[1].set_xlim(axs[0].get_xlim())
plt.show()
```
